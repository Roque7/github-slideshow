```python
import math
def extract(x):
    temp = x.split(" ")
    if  "Mr." in temp:
        return "Mr"
    elif "Mrs." in temp:
        return "Mrs"
    elif "Miss." in temp:
        return "Miss"
    elif "Master." in temp:
        return "Master"
    elif "Dr." in temp:
        return "Dr"
    else:
        return None
    
df["Category"] = df["Name"].apply(extract)

df.head()
```

```python
import numpy as np
def fillAge(row):
    if(math.isnan(row.Age)):
        if row.Category == "Mr":
            return np.mean(df[df["Category"] == "Mr"]["Age"])
        elif row.Category == "Mrs":
            return np.mean(df[df["Category"] == "Mrs"]["Age"])
        elif row.Category == "Miss":
            return np.mean(df[df["Category"] == "Miss"]["Age"])
        elif row.Category == "Master":
            return np.mean(df[df["Category"] == "Master"]["Age"])
        elif row.Category == "Dr":
            return np.mean(df[df["Category"] == "Dr"]["Age"])
        else:
            return None
    else:
        return row.Age

df["Age"] = df.apply(lambda row: fillAge(row), axis = 1)
df.info()
```

```python
q = 0.05
qMaxima = df["Age"].quantile(1-q)
qMinima = df["Age"].quantile(q)
```

```python
dfSinOutliers = df[(df["Age"] < qMaxima) & (df["Age"] > qMinima)]
sns.boxplot(x=dfSinOutliers["Age"])
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df["EmbarkedClass"] = le.fit_transform(df["Embarked"])
df.head()
```

```python
df["SexClass"] = le.fit_transform(df["Sex"])
df.head()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
escaler = MinMaxScaler()
df[["FareScaled"]] = escaler.fit_transform(df[["Fare"]])
df[["AgeScaled"]] = escaler.fit_transform(df[["Age"]])
df[["SexClassScaled"]] = escaler.fit_transform(df[["SexClass"]])
df.head()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
#Como se ha comentado anteriormente, tener muchas clases distintas no nos favorece necesariamente. Ante esta situación podemos hacer grupos para poder reducir el número de clases, por ejemplo por rango de edades.
df['Age-bin'] = df['Age']
df.loc[df['Age-bin'] <= 16, 'Age-bin'] = 0
df.loc[(df['Age-bin'] > 16) & (df['Age-bin'] <= 32), 'Age-bin'] = 1
df.loc[(df['Age-bin'] > 32), 'Age-bin'] = 2
#Un aspecto a añadir pueden ser los **dummies** en los cuales en lugar de tener una columna donde aparecen múltiples clases tenemos una columna con cada una de las clases indicando si el elemento de esa fila pertence a la clase '1' o no '0'.
df['FamilySize'] = df['Parch'] + df['SibSp'] + 1
df['Singleton'] = df['FamilySize'].map(lambda s: 1 if s == 1 else 0)
df['SmallFamily'] = df['FamilySize'].map(lambda s: 1 if 2 <= s <= 3 else 0)
df['MediumFamily'] = df['FamilySize'].map(lambda s: 1 if s == 4 else 0)
df['LargeFamily'] = df['FamilySize'].map(lambda s: 1 if 5 <= s else 0)
```
